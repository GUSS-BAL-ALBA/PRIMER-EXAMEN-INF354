### **Primera tecnica**:

**fill**: propaga la última observación válida a la siguiente validación.
  **backfill / bfill**: utilice la siguiente observación válida para llenar el vacío.

  Es un de las tecnicas que util a cuando no se quiere manejar funciones estadisticas como la media,

```python
#    ffill:              propagate last valid observation forward to next valid.
#    backfill / bfill:   use next valid observation to fill gap.

df_fill = df.fillna(method ='bfill')
df_fill
```

### **tercera tecnica:**

es una extencion de la segunda tecnica.

donde se crea un dicionario que en las claves del diccionario estan las columnas, y como value van los valores que quisieramos que remplace a los datos vacios.

```python
fd_areglado=df
valores_por_defecto={"Price_CLP":fd_areglado['Price_CLP'].fillna(df['Price_CLP'].mean()),
                     "Price_UF":fd_areglado['Price_UF'].fillna(df['Price_UF'].mode()),
                     "Price_USD":fd_areglado['Price_USD'].fillna(df['Price_USD'].mean()),
                     "Comuna":"DESCONOCIDO",
                     "Ubicacion":"DESCONOCIDO",
                     "Dorms":fd_areglado['Price_UF'].fillna(df['Price_UF'].mode()),
                     "Baths":fd_areglado['Price_UF'].fillna(df['Price_UF'].mode()),
                     "Built Area":"DESCONOCIDO",
                     "Total Area":fd_areglado['Price_CLP'].fillna(df['Price_CLP'].mean()),
                     "Parking":fd_areglado['Price_CLP'].fillna(df['Price_CLP'].mean()),
                     "id":fd_areglado['Price_CLP'].fillna(df['Price_CLP'].max()+1),
                     "Realtor":"DESCONOCIDO",
                     "Comuna_categoria":fd_areglado['Price_CLP'].fillna(df['Price_CLP'].mean())

                     }
df_sin_nulos=fd_areglado.fillna(value=valores_por_defecto)
print(df_sin_nulos)
```

### **exportar dataSet sin valores nulos**

```python
df_sin_nulos.to_csv("data_casas.csv")
```
