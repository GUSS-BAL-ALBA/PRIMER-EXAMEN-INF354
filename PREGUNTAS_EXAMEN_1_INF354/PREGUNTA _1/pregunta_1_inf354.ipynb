```python
import _csv
```

```python
lista=[]

with open('/content/drive/MyDrive/inf-354/datos/2023-03-08 Precios Casas RM.csv','r',encoding='utf-8') as data:
  entrada=_csv.reader(data,delimiter=',')
  print(next(entrada))
  #next(entrada)
  lista=list(entrada)
```

```python
media=0
moda=0
lista_dolares=[]
```

calculamos la **media**  media=sumatoria(xi)/nro_elementos


tomamos como referencia la columna de **Precio en dolares de las casas**

```python
for linea in lista:
  lista_dolares.append(eval(linea[2]))
  media=media + eval(linea[2])
media=media/len(lista)
print("La media es:  "+str(media)+"  en Dolares")
```

Calculamos la **Moda**

```python
lista_moda={}

for linea in lista:
    clave = str(eval(linea[2]))
    # Si no existe...
    if not clave in lista_moda:
        # lo agregamos:
        lista_moda[clave] = 1
    # Si ya existe...
    else:
        # Lo aumentamos
        lista_moda[clave] += 1
```

```python
print(lista_moda)
```

```python
moda=[llave for llave, valor in lista_moda.items() if valor == max(lista_moda.values())]
print( "El valor  que mas se repite(MODA) es:  "+str(moda))
```
